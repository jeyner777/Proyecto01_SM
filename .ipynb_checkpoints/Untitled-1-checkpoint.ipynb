```python

def standardize_date(date_series):
    standardized_dates = pd.to_datetime(date_series, errors='coerce')
    
    mask = standardized_dates.isna()
    if mask.any():
        retry_dates = pd.to_datetime(date_series[mask], format='%d/%m/%Y', errors='coerce')
        standardized_dates[mask] = retry_dates
    
    return standardized_dates.dt.date

def standardize_time(time_series):
    standardized_times = pd.to_datetime(time_series, format='%H:%M:%S', errors='coerce').dt.time
    
    mask = standardized_times.isna()
    if mask.any():
        retry_times = pd.to_datetime(time_series[mask], format='%H:%M:%S.%f', errors='coerce').dt.time
        standardized_times[mask] = retry_times
    
    return standardized_times

combined_csv['Fecha_Retiro'] = standardize_date(combined_csv['Fecha_Retiro'])
combined_csv['Fecha_Arribo'] = standardize_date(combined_csv['Fecha_Arribo'])
combined_csv['Hora_Retiro'] = standardize_time(combined_csv['Hora_Retiro'])
combined_csv['Hora_Arribo'] = standardize_time(combined_csv['Hora_Arribo'])

print(combined_csv.head())

combined_csv.info()
```

```python

def safe_time_conversion(time_series):
    try:
        # Intenta primero con un formato que incluye milisegundos
        return pd.to_datetime(time_series, format='%H:%M:%S.%f', exact=False).dt.time
    except ValueError:
        # Si falla, intenta con un formato sin milisegundos
        return pd.to_datetime(time_series, format='%H:%M:%S', exact=False).dt.time

# Convertir las columnas de hora
combined_csv['Hora_Retiro'] = safe_time_conversion(combined_csv['Hora_Retiro'])
combined_csv['Hora_Arribo'] = safe_time_conversion(combined_csv['Hora_Arribo'])

# Verificando los tipos de datos
combined_csv.dtypes
```
